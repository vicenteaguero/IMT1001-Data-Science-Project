```python
import numpy as np #quizás sobren algunas cosas de acá
import matplotlib.image as img
from skimage import io
from PIL import Image
import cv2
```

```python
def conversor(imagen):
    col = Image.open(imagen)
    bw = col.convert('L') #convierte la imagen a blanco y negro
    bw.save("bw_image.jpg") 
    img_array = cv2.imread("bw_image.jpg", cv2.IMREAD_GRAYSCALE)
    img_array = cv2.bitwise_not(img_array)

    img_size=28    #transforma el tamaño
    new_array = cv2.resize(img_array, (img_size,img_size)) 

    final_array = []
    for lista in new_array:    #cambia un poco el formato q da lo de arriba para q coincida con el q ocupamos
        for numero in lista:
            final_array.append(float(numero))
    final_array = np.array(final_array) 
    final_array= final_array.reshape(1, -1) #array del formato necesario

    return(final_array)
```

```python
base_de_datos = [] #acá debería quedar guardado todo, no estoy tan segura si sirve una lista eso sí
```

```python
for i in range (3): #esto es para no tener q agregarlo uno por uno a la base de datos, el range es la cantidad de fotos
    matriz = conversor("numero_{}.png".format(i))#la idea es ponerle los nombres a las fotos con este formato "numero_X"
    base_de_datos.append(matriz)
```
